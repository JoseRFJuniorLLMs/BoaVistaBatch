# **Problema**

O presente problema se refere aos dados de uma empresa que produz e vende mquinas
industriais. O objetivo deste desafio  

> (1) Fazer a modelagem conceitual dos dados.

> (2) Criar a infraestrutura necessria.

> (3) Criar todos os artefatos necessrios para carregar os arquivos para o banco criado.

Os seguintes arquivos devem ser importados para o banco de dados:

 > price_quote.csv: Esse arquivo contm informaes sobre price quotes dos fornecedores. Os preos podem ser cotados de 2 formas:

>> 1. Bracket pricing: possui mltiplos nveis de compra, baseado em quantidades (isso significa que o custo  dado assumindo-se a compra de certa quantidade de tubos) 

>> 2. Non- bracket pricing: possui uma quantidade mnima a ser pedida (min_order) para que o preo seja aplicado.
Cada uma dessas cotaes  emitida com uma annual_usage, que se refere a estimativa de quantos conjuntos de tubos sero comprados em determinado ano.

 > Bill_of_materials.csv: Esse arquivo contm a lista de componentes e as quantidades que foram utilizadas em cada conjunto de tubos.

 > comp_boss.csv: Esse arquivo contm as informaes sobre cada componente.

Dicionrio de Dados

Nome do campo.
Chave.
A chave Primria (P)  uma coluna ou conjunto de colunas que identifica exclusivamente o restante dos dados em qualquer linha especfica da tabela.
A chave Externa (E)  uma coluna na tabela que  uma chave primria em outra tabela, indicando que quaisquer dados em uma coluna de chave externa deve ter dados correspondentes na outra tabela em que essa coluna  a chave primria. Para bancos de dados, essa correspondncia  conhecida como integridade referencial.
O valor NOT NULL (NN) significa que a coluna deve ter um valor em cada linha. Se NULL foi utilizado, essa coluna pode ser deixada em branco em uma linha especfica.
